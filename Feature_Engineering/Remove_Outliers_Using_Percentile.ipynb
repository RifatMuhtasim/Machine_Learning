```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Remove Outliers from Housing Dataset

```python
df0 = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/FeatureEngineering/1_outliers/bhp.csv")

print("Shape:", df0.shape)
df0.head()
```

```python
df0.describe()
```

```python
# Show the value of the quantile
df0['price_per_sqft'].quantile(0.95)
```

```python
min_threshold, max_threshold = df0['price_per_sqft'].quantile([0.01, 0.99])

print("Min Threshold:", min_threshold)
print("Max Threshold:", max_threshold)
```

```python
# Outliers who are below the min threshold value
df0[df0['price_per_sqft'] < min_threshold]
```

```python
# Outliers which are the greater then the max threshold value
df0[df0['price_per_sqft'] > max_threshold]
```

```python
df = df0[(df0['price_per_sqft'] > min_threshold) & (df0['price_per_sqft'] < max_threshold) ]

print("Shape:", df.shape)
df.head()
```

```python
df.describe()
```

```python
df.sample(10)
```

# Remove Outliers from Airbnb Dataset

```python
airbnb_df0 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Machine_Learning/Feature_Engineering/AB_NYC_2019.csv")

print("Shape:", airbnb_df0.shape)
airbnb_df0.head()
```

```python
airbnb_df0.describe()
```

```python
min_threshold, max_threshold = airbnb_df0['price'].quantile([0.01, 0.999])

print("Min Threshold:", min_threshold)
print("Max Threshold:", max_threshold)
```

```python
# Minimum Outliers
airbnb_df0[airbnb_df0['price'] < min_threshold].shape
```

```python
# Maximum Outliers
airbnb_df0[airbnb_df0['price'] > max_threshold].shape
```

```python
airbnb_df = airbnb_df0[(airbnb_df0['price'] > min_threshold) & (airbnb_df0['price'] < max_threshold) ]

print("Shape:", airbnb_df.shape)
airbnb_df.head()
```

```python
result = airbnb_df0.shape[0] - airbnb_df.shape[0]
print("Data Remove:", result)
```
