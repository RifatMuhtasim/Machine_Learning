```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# Data Preprocessing

```python
# Read dataset
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/18_PCA/Exercise/heart.csv")

print("Shape:", df.shape)
df.head()
```

```python
df.corr()
```

```python
df.isna().sum()
```

```python
df.nunique()
```

# Remove Outliers using Z Score.

```python
# Check the RestingBP outliers
df[df['RestingBP'] > ( df['RestingBP'].mean() + 3*df['RestingBP'].std() )]
```

```python
df = df[df['RestingBP'] <= ( df['RestingBP'].mean() + 3*df['RestingBP'].std() )]
```

```python
# Check the Cholesterol Outliers
df[df['Cholesterol'] > ( df['Cholesterol'].mean() + 3*df['Cholesterol'].std() )]
```

```python
df = df[df['Cholesterol'] <= ( df['Cholesterol'].mean() + 3*df['Cholesterol'].std() )]
```

```python
# Check the MaxHR Outliers
df[df['MaxHR'] > ( df['MaxHR'].mean() + 3*df['MaxHR'].std() )]
```

```python
# Check the Oldpeak Outliers
df[df['Oldpeak'] > ( df['Oldpeak'].mean() + 3*df['Oldpeak'].std() )]
```

```python
df = df[df['Oldpeak'] <= ( df['Oldpeak'].mean() + 3*df['Oldpeak'].std() )]
```

```python
print("Shape:", df.shape)
df.head()
```

# Convert Text Columns into Numbers

```python
print('RestingECG: ', df['RestingECG'].unique())
print('ExerciseAngina: ', df['ExerciseAngina'].unique())
print('ST_Slope: ', df['ST_Slope'].unique())
```

```python
df['RestingECG'].replace({
    "Normal": 1,
    "ST": 2,
    "LVH": 3
}, inplace=True)

df['ExerciseAngina'].replace({
    "N": 0,
    "Y": 1
}, inplace=True)

df['ST_Slope'].replace({
    "Down": 1,
    "Flat": 2,
    "Up": 3
}, inplace=True)

df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

# Apply Scaling

```python
X = df.drop('HeartDisease', axis="columns")
y = df['HeartDisease']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
print(X_scaler)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

# Model

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_model.score(X_test, y_test)
```

```python
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_model.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_model.score(X_test, y_test)
```

**Comment:** Random Forest is best

# Use PCA to reduce dimensions

```python
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
print(X_pca)
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.25)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train_pca, y_train)
rf.score(X_test_pca, y_test)
```
