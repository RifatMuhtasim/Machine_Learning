**PCA:** Principal Component Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# Data Preprocessing

```python
# Read Dataset
from sklearn.datasets import load_digits
digits = load_digits()
dir(digits)
```

```python
value = 67
print("Value:", digits['target'][value])

plt.gray()
plt.matshow(digits['images'][value])
plt.show()
```

```python
df = pd.DataFrame(digits['data'], columns=digits['feature_names'])
df.head()
```

```python
df.describe()
```

# Train Test Split

```python
X = df
y = digits['target']
```

```python
# Transform this X value with StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
```

```python
# Train Test Split

X_train, X_test, y_train, y_test = train_test_split(X_scaler, y, test_size = 0.25)
print("X Train len:", len(X_train))
print("X Test len:", len(X_test))
```

```python
lm_model = LogisticRegression()
lm_model.fit(X_train, y_train)
lm_model.score(X_test, y_test)
```

# Use PCA to reduce dimensions

```python
X.head()
```

```python
# Use components such that 95% of variance is retained

from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
# Divided the result with Train Test Split

X_pca_train, X_pca_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.25)
```

```python
model = LogisticRegression()
model.fit(X_pca_train, y_train)
model.score(X_pca_test, y_test)
```
