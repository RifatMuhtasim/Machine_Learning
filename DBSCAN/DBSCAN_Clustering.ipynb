# DBSCAN <br>
It is helpful when non linear dataset. <br>
Parameter: minPts = 4 and $\epsilon =$ radius. <br>
It has three points:
1. Core Points: Number of points within the radius should be grater than or equal minPts.
2. Border Points: No data points within the radius will be less than minPts = 4
3. Noise/ Outliers Points: Only one number data points inside the radius. Which is Robust to the outliers.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/a/af/DBSCAN-Illustration.svg/1024px-DBSCAN-Illustration.svg.png"  style="width: 50%;"/>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

import warnings
warnings.filterwarnings("ignore")
```

```python
X, _ = make_moons(n_samples=250, noise=0.05)
X[:10]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=_)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled[:10]
```

# DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5)
dbscan.fit(X_scaled)

y_labels = dbscan.labels_
y_labels[:10]
```

```python
y_labels_series = pd.Series(y_labels)
y_labels_series.value_counts()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_labels)
plt.show()
```
