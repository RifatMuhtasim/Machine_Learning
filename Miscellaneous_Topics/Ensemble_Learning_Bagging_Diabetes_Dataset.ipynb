```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data Preprocessing

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/19_Bagging/diabetes.csv")
print(df.shape)
df.head()
```

```python
# Check the null value
df.isnull().sum()
```

```python
# DataFrame Summary
df.describe()
```

# Scaling the dataset

```python
X = df.drop('Outcome', axis="columns")
y = df['Outcome']
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
X_scaler
```

# Split the dataset into Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaler, y, test_size = 0.20)
```

```python
print("X_train length:", X_train.shape)
print("X_test length:", X_test.shape)
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)
```

```python
tree_model.score(X_test, y_test)
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_score = cross_val_score(DecisionTreeClassifier(), X, y, cv=5)
cross_score
```

```python
cross_score.mean()
```

# Ensemble - Bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_model = BaggingClassifier(
    estimator= DecisionTreeClassifier(),
    n_estimators = 100,
    max_samples = 0.8,
    oob_score = True,
    random_state = 0
)

bag_model.fit(X_train, y_train)
bag_model.oob_score_
```

## Bagging Classifier with cross_val_score

```python
bag_model = BaggingClassifier(
    estimator = DecisionTreeClassifier(),
    n_estimators = 100,
    max_samples = 0.8,
    oob_score = True,
    random_state = 0
)

scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```
